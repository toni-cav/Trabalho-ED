```python
df_compras = df.where(df['Type']=='C').dropna()
df_compras.count()
df_compras
```

```python
df_vendas = df.where(df['Type']=='V').dropna()
df_vendas.count()
df_vendas
```

```python
lista_venda = Linked_List_Venda()
IDs_vendas = df_vendas['ID'].to_list()

inicio = time.time()

for i in IDs_vendas: # df['ID] é uma unidade maior que os índices do df. Por isso usamos i-1
    ordem1 = Ordem(
        id_ordem = df_vendas.loc[i-1, 'ID'],
        tipo = df_vendas.loc[i-1, 'Type'],
        preco = df_vendas.loc[i-1, 'Price'],
        quantidade = df_vendas.loc[i-1, 'Quantity']
    )
    acao1 = Node(ordem1)
    
    lista_venda.adicionar_no(acao1)

fim = time.time()

tempo = fim - inicio
print(tempo)
```

```python
lista_compra = Linked_List_Compra()
IDs_compras = df_compras['ID'].to_list()

inicio = time.time()

for i in IDs_compras: # df['ID] é uma unidade maior que os índices do df. Por isso usamos i-1



    ordem1 = Ordem(
        id_ordem = df_compras.loc[i-1, 'ID'],
        tipo = df_compras.loc[i-1, 'Type'],
        preco = df_compras.loc[i-1, 'Price'],
        quantidade = df_compras.loc[i-1, 'Quantity']
    )
    acao1 = Node(ordem1)
    
    lista_compra.adicionar_no(acao1)

fim = time.time()

tempo_execucao = fim - inicio

print(tempo_execucao)
```

```python
pilha_compra = Stack()


inicio = time.time()

for i in IDs_compras: # df['ID] é uma unidade maior que os índices do df. Por isso usamos i-1
    ordem1 = Ordem(
        id_ordem = df_compras.loc[i-1, 'ID'],
        tipo = df_compras.loc[i-1, 'Type'],
        preco = df_compras.loc[i-1, 'Price'],
        quantidade = df_compras.loc[i-1, 'Quantity']
    )
    acao1 = Node(ordem1)
    
    pilha_compra.empilhar(acao1)

fim = time.time()

tempo_execucao = fim - inicio

print('Tempo para empilhar')
print(tempo_execucao)

inicio = time.time()
pilha_compra.desempilhar()
fim = time.time()

print('\nTempo para desimpilhar')
print(fim-inicio)
```

```python
queue_compra = Queue()


inicio = time.time()
for i in IDs_compras: # df['ID] é uma unidade maior que os índices do df. Por isso usamos i-1
    ordem1 = Ordem(
        id_ordem = df_compras.loc[i-1, 'ID'],
        tipo = df_compras.loc[i-1, 'Type'],
        preco = df_compras.loc[i-1, 'Price'],
        quantidade = df_compras.loc[i-1, 'Quantity']
    )
    acao1 = Node(ordem1)
    
    queue_compra.enqueue(acao1)

fim = time.time()

tempo_execucao = fim - inicio

print('Tempo para enqueue')
print(tempo_execucao)

inicio = time.time()
queue_compra.dequeue
fim = time.time()

print('\nTempo para dequeue')
print(fim-inicio)
```
